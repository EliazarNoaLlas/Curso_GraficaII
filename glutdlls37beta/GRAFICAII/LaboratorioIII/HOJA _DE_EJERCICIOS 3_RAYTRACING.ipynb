```python
import numpy as np
import matplotlib.pyplot as plt

def normalize_vector(vector):
    return vector / np.linalg.norm(vector)

def cylinder_intersection(center, height, radius, ray_origin, ray_direction):
    EPSILON = 0.0000001

    ray_to_center = center - ray_origin
    proj = np.dot(ray_direction, np.cross(ray_to_center, ray_direction))
    dist_to_center = np.linalg.norm(ray_to_center)
    dist_to_axis = np.linalg.norm(proj)
    dist_to_caps = np.abs(dist_to_center ** 2 - dist_to_axis ** 2) ** 0.5

    if dist_to_axis > radius or dist_to_caps > height:
        return None

    # Verificar si los valores dentro de la raz cuadrada son no negativos
    value_inside_sqrt = radius ** 2 - dist_to_caps ** 2
    if value_inside_sqrt < 0:
        return None

    t = dist_to_axis - value_inside_sqrt ** 0.5

    return t


def find_nearest_intersected_object(objects, ray_origin, ray_direction):
    distances = [cylinder_intersection(obj['center'], obj['height'], obj['radius'], ray_origin, ray_direction) for obj in objects]
    nearest_object = None
    min_distance = np.inf
    for index, distance in enumerate(distances):
        if distance and distance < min_distance:
            min_distance = distance
            nearest_object = objects[index]
    return nearest_object, min_distance

width = 300
height = 200

camera = np.array([0, 0, 1])
aspect_ratio = float(width) / height
screen = (-1, 1 / aspect_ratio, 1, -1 / aspect_ratio)

objects = [
    {'center': np.array([0, 0, -1]),
     'height': 2,
     'radius': 0.5,
     'color': [255, 0, 0]},
    {'center': np.array([1, -0.5, -2]),
     'height': 1.5,
     'radius': 0.3,
     'color': [0, 255, 0]},
]

image = np.zeros((height, width, 3))
for i, y in enumerate(np.linspace(screen[1], screen[3], height)):
    for j, x in enumerate(np.linspace(screen[0], screen[2], width)):
        pixel = np.array([x, y, 0])
        ray_origin = camera
        ray_direction = normalize_vector(pixel - ray_origin)
        nearest_object, min_distance = find_nearest_intersected_object(objects, ray_origin, ray_direction)
        if nearest_object is None:
            continue
        intersection = ray_origin + min_distance * ray_direction
        color = nearest_object['color']
        image[i, j] = np.clip(color, 0, 1)
        print("%d/%d" % (i + 1, height))

plt.imsave('imagen.png', image)
imgplot = plt.imshow(image)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def normalizar(vector):
    return vector / np.linalg.norm(vector)

def interseccion_cilindro(centro, radio, altura, origen_rayo, direccion_rayo):
    a = direccion_rayo[0] ** 2 + direccion_rayo[1] ** 2
    b = 2 * (direccion_rayo[0] * (origen_rayo[0] - centro[0]) + direccion_rayo[1] * (origen_rayo[1] - centro[1]))
    c = (origen_rayo[0] - centro[0]) ** 2 + (origen_rayo[1] - centro[1]) ** 2 - radio ** 2
    discriminante = b ** 2 - 4 * a * c

    if discriminante >= 0:
        t1 = (-b + np.sqrt(discriminante)) / (2 * a)
        t2 = (-b - np.sqrt(discriminante)) / (2 * a)

        t_min = min(t1, t2)
        t_max = max(t1, t2)

        # Comprobar si la interseccin ocurre dentro de la altura del cilindro
        z_min = centro[2]
        z_max = centro[2] + altura

        if t_min <= z_min and z_min <= t_max:
            return z_min
        elif t_min <= z_max and z_max <= t_max:
            return z_max

    return None


def objeto_intersectado_mas_cercano(objetos, origen_rayo, direccion_rayo):
    distancias = []
    for obj in objetos:
        if obj['tipo'] == 'cilindro':
            distancia = interseccion_cilindro(obj['centro'], obj['radio'], obj['altura'], origen_rayo, direccion_rayo)
        elif obj['tipo'] == 'triangulo':
            distancia = interseccion_triangulo(obj['vertice0'], obj['vertice1'], obj['vertice2'], origen_rayo, direccion_rayo)
        else:
            distancia = None
        distancias.append(distancia)

    objeto_mas_cercano = None
    distancia_minima = np.inf
    for indice, distancia in enumerate(distancias):
        if distancia and distancia < distancia_minima:
            distancia_minima = distancia
            objeto_mas_cercano = objetos[indice]

    return objeto_mas_cercano, distancia_minima

ancho = 300
alto = 200

camara = np.array([0, 0, 2])
proporcion = float(ancho) / alto
pantalla = (-1, 1 / proporcion, 1, -1 / proporcion)

objetos = [
    { 'tipo': 'cilindro', 'centro': np.array([-0.5, 0, -1]), 'radio': 0.2, 'altura': 1.5, 'color': [255, 0, 0] },
    { 'tipo': 'triangulo', 'vertice0': np.array([-0.5, -0.5, -1]), 'vertice1': np.array([0.5, -0.5, -1]), 'vertice2': np.array([0, 0.5, -1]), 'color': [0, 255, 0] },
    # Agrega ms objetos segn tus necesidades
]

imagen = np.zeros((alto, ancho, 3))
for i, y in enumerate(np.linspace(pantalla[1], pantalla[3], alto)):
    for j, x in enumerate(np.linspace(pantalla[0], pantalla[2], ancho)):
        pixel = np.array([x, y, 0])
        origen = camara
        direccion = normalizar(pixel - origen)

        objeto_mas_cercano, distancia_minima = objeto_intersectado_mas_cercano(objetos, origen, direccion)
        if objeto_mas_cercano is None:
            continue

        interseccion = origen + distancia_minima * direccion
        color = objeto_mas_cercano['color']
        imagen[i, j] = np.clip(color, 0, 255)
imagen = imagen.astype(np.float32) / 255.0  # Normalizar los valores de color en el rango de 0 a 1

plt.imsave('imagen.png', imagen)
imgplot = plt.imshow(imagen)
plt.show()
```
